```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import geodatasets

from scipy import sparse
from scipy.sparse import hstack
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, auc, roc_curve
from nltk.corpus import stopwords
from collections import Counter
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv("fake_job_postings.csv")

# Combine relevant text fields
text_cols = ['title', 'company_profile', 'description', 'requirements', 'benefits']
df[text_cols] = df[text_cols].fillna('')
df['text'] = df[text_cols].apply(lambda x: ' '.join(x), axis=1)

# Target
df['target'] = df['fraudulent']
```

```python
# EDA

# Set up plotting aesthetics
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Clean a few key categorical variables
df['fraudulent'] = df['fraudulent'].astype('category')
df['telecommuting'] = df['telecommuting'].astype('category')
df['has_company_logo'] = df['has_company_logo'].astype('category')

# Plot: Distribution of fraudulent vs real job postings
sns.countplot(data=df, x='fraudulent')
plt.title("Distribution of Fraudulent vs Real Job Postings")
plt.xlabel("Is Fraudulent?")
plt.ylabel("Count")
plt.show()

# Plot: Employment type distribution by fraud status
sns.countplot(data=df, y='employment_type', hue='fraudulent')
plt.title("Employment Type by Fraud Status")
plt.xlabel("Count")
plt.ylabel("Employment Type")
plt.legend(title="Fraudulent")
plt.show()

# Plot: Top 10 most common industries
top_industries = df['industry'].value_counts().nlargest(10)
sns.barplot(x=top_industries.values, y=top_industries.index)
plt.title("Top 10 Most Common Industries in Job Postings")
plt.xlabel("Count")
plt.ylabel("Industry")
plt.show()

# Extract the US state abbreviation from the location string
df['state'] = df['location'].astype(str).str.extract(r'US,\s*([A-Z]{2})')

# Count real and fake postings by state
state_fraud_counts = df.groupby(['state', 'fraudulent'], observed=True).size().unstack(fill_value=0)
state_fraud_counts.columns = ['Real', 'Fake']
state_fraud_counts['Total'] = state_fraud_counts['Real'] + state_fraud_counts['Fake']
state_fraud_counts['Fake_Proportion'] = state_fraud_counts['Fake'] / state_fraud_counts['Total']
state_fraud_counts['Real_Proportion'] = state_fraud_counts['Real'] / state_fraud_counts['Total']

# Top 10 states by fake proportion (min 10 postings)
top10_fake_prop = state_fraud_counts[state_fraud_counts['Total'] >= 10].sort_values(by='Fake_Proportion', ascending=False).head(10).reset_index()

# Top 10 states by real proportion
top10_real_prop = state_fraud_counts[state_fraud_counts['Total'] >= 10].sort_values(by='Real_Proportion', ascending=False).head(10).reset_index()

# Add dummy hue for future seaborn compatibility
top10_fake_prop['hue'] = top10_fake_prop['state']
top10_real_prop['hue'] = top10_real_prop['state']

# Plot side-by-side proportion comparison
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Left: Fake job proportion
sns.barplot(x='Fake_Proportion', y='state', data=top10_fake_prop, color='firebrick', ax=axes[0])
axes[0].set_title("Top 10 States by Fake Job Proportion", fontsize=14)
axes[0].set_xlabel("Proportion of Fake Postings")
axes[0].set_ylabel("State")

# Right: Real job proportion
sns.barplot(x='Real_Proportion', y='state', data=top10_real_prop, color='forestgreen', ax=axes[1])
axes[1].set_title("Top 10 States by Real Job Proportion", fontsize=14)
axes[1].set_xlabel("Proportion of Real Postings")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()

# Correlation Matrix: Numeric Features vs Fraudulent
df_corr = df[['fraudulent', 'telecommuting', 'has_company_logo', 'has_questions']].copy()
df_corr['fraudulent'] = df_corr['fraudulent'].astype(int)

corr_matrix = df_corr.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Matrix with Fraudulent Indicator")
plt.tight_layout()
plt.show()
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words("english")) - {"not", "no", "nor"}

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df['text'] = df['text'].apply(preprocess_text)
```

```python
# TF-IDF (sparse)
tfidf = TfidfVectorizer(max_features=5000)
X_text = tfidf.fit_transform(df["text"]) 

# Numeric features
numeric_cols = ['telecommuting', 'has_company_logo', 'has_questions']
X_numeric = df[numeric_cols].astype(float).values

# Scale numeric features to [0, 1]
scaler = MinMaxScaler()
X_numeric_scaled = scaler.fit_transform(X_numeric)

# Convert numeric to sparse
X_numeric_sparse = sparse.csr_matrix(X_numeric_scaled)

# Combine text + numeric (sparse + sparse = sparse)
X_combined = hstack([X_text, X_numeric_sparse])

# Target
y = df["target"]
```

```python
selector = SelectKBest(score_func=chi2, k=1000)
X_selected = selector.fit_transform(X_combined, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Define and train models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, solver='liblinear'),
    "Naive Bayes": MultinomialNB(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Accuracy
    acc = accuracy_score(y_test, y_pred)

    # Confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

    # Specificity = TN / (TN + FP)
    specificity = tn / (tn + fp)

    # Sensitivity = TP / (TP + FN)
    sensitivity = tp / (tp + fn)

    # Store results
    results[name] = {
        "Accuracy": round(acc, 4),
        "Specificity": round(specificity, 4),
        "Sensitivity (Recall)": round(sensitivity, 4)
    }

    # Print full report
    print(f"\n=== {name} ===")
    print(f"Accuracy: {acc:.4f}")
    print(f"Specificity (True Negative Rate): {specificity:.4f}")
    print(f"Sensitivity (Recall for Fakes): {sensitivity:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
# Convert to dataframe
results_df = pd.DataFrame(results).T

# Display sorted by Accuracy
print("\nModel Performance Summary:")
print(results_df.sort_values("Accuracy", ascending=False))
```

```python
# Reset index for plotting
plot_df = results_df.reset_index().rename(columns={"index": "Model"})

plt.figure(figsize=(10, 6))
sns.barplot(data=plot_df, x="Accuracy", y="Model", palette="viridis")
plt.title("Model Accuracy Comparison")
plt.xlim(0.8, 1.0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=plot_df, x="Specificity", y="Sensitivity (Recall)", hue="Model", s=100)
plt.title("Specificity vs Sensitivity by Model")
plt.xlabel("Specificity (True Negative Rate)")
plt.ylabel("Sensitivity (Recall for Fake Jobs)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Get predicted probabilities for the positive class (label = 1)
rf_model = models["Random Forest"]
y_probs = rf_model.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Random Forest (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curve: Random Forest")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Word Clouds
real_text = " ".join(df[df["target"] == 0]["text"])
fake_text = " ".join(df[df["target"] == 1]["text"])

real_wc = WordCloud(width=800, height=400, background_color="white").generate(real_text)
fake_wc = WordCloud(width=800, height=400, background_color="black", colormap="Reds").generate(fake_text)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.imshow(real_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud - Real Jobs")

plt.subplot(1, 2, 2)
plt.imshow(fake_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud - Fake Jobs")

plt.tight_layout()
plt.show()

#Barplots
# Word frequency function
def get_word_frequencies(text):
    words = text.split()
    return Counter(words)

real_word_freq = get_word_frequencies(real_text)
fake_word_freq = get_word_frequencies(fake_text)

real_top_20 = real_word_freq.most_common(20)
fake_top_20 = fake_word_freq.most_common(20)

# Plot
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
words, counts = zip(*real_top_20)
sns.barplot(x=list(counts), y=list(words), hue=list(words), palette="Greens_r", dodge=False, legend=False)
plt.title("Top 20 Words in Real Job Postings")
plt.xlabel("Frequency")
plt.ylabel("Words")

plt.subplot(1, 2, 2)
words, counts = zip(*fake_top_20)
sns.barplot(x=list(counts), y=list(words), hue=list(words), palette="Reds_r", dodge=False, legend=False)
plt.title("Top 20 Words in Fake Job Postings")
plt.xlabel("Frequency")
plt.ylabel("Words")

plt.tight_layout()
plt.savefig("top_words_real_vs_fake.png", dpi=300, bbox_inches='tight')
plt.show()
```
